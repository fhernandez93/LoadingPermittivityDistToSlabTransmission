```python
#Regular plots

folder_path = ["output/Convergence_STL_Test_Sliced_phi_27_const_perm/z_incidence/Data"
               ]
names = ["$\\text{Bandgap evolution as a function of thickness }\phi = 0.27 \\text{ - }\epsilon=11.56$"]
edge = ["1/10","2/10","3/10","4/10","5/10","6/10","7/10","8/10","9/10"]
for j,item in enumerate(folder_path):
 i =0
 fig = go.Figure()
 for filename in sorted(os.listdir(item)):
    print(filename)
    if not Path(filename).suffix==".h5":
        continue 
    if os.path.isfile(os.path.join(item, filename)):
        file=os.path.join(item, filename)
        # Load HDF5 file
        with h5py.File(file, 'r') as f:
            lambdas = 0.34/np.array(f['lambdas'])
            transmission = np.array(f['transmission_right'])
            transmission_left = np.array(f['transmission_left'])

            
        # Add data to the plot
        fig.add_trace(go.Scatter(x=lambdas, y=transmission,
                                    mode='lines'
                                    ,name=f"L'={edge[i]}L"
                            ))
        i+=1
      
  # Update layout
 fig.update_yaxes(type="log")
 fig.update_layout(
                title=dict(text=names[j]),
                xaxis_title=r'$\Large{a/\lambda}$',
                yaxis_title=r'$\large{\text{T}}$',
                #yaxis_range=[-5.4,0],  # Set the range for y-axis
                yaxis_gridcolor='lightgrey',  # Set grid color
                plot_bgcolor='rgba(0,0,0,0)',  # Set background color of plot,
                font=dict(size=20),
                yaxis = dict(
                    showexponent = 'all',
                    exponentformat = 'e'
                    )
                
                )


#Save the plot in PDF format
 store_path = f"output/Plots_STL/Poster/Slices"
 if not os.path.exists(store_path):
       os.makedirs(store_path)
       print(f"Folder '{store_path}' created successfully.")
     
 fig.write_image(f"{store_path}/slices_phi_27.png",format="png", engine="kaleido")
 fig.show()
        
```

```python
#Gap position plots
# Load the CSV file provided by the user
file_path = 'distributions/Schinke.csv'
data_refractive = pd.read_csv(file_path)
data_abs= pd.read_csv('distributions/Schinke copy.csv')

folder_path = ["output/Convergence_STL_Test_phi_Absorption/z_incidence/Data"]
sizes = [0.596,0.616,0.662,0.715,0.812,0.894,1.19,1.78]
transmission_mins=[]
lambdas_mins=[]
for j,item in enumerate(folder_path):
 i =0
 fig = make_subplots(rows=2, cols=1,shared_xaxes=True,subplot_titles=(r"$\Large{\text{Si refractive index/absorption}}$",r"$\Large{\text{Gap Position vs } \lambda}$"))

 for filename in sorted(os.listdir(item)):
    print(filename)
    if filename == "test_network_true_size1.7875_size_1.79_runtime_120_lambdaRange_0.95-0.45_incidence_z_decay_0.0249.h5":
        continue
    if not Path(filename).suffix==".h5":
        continue 
    if os.path.isfile(os.path.join(item, filename)):
        file=os.path.join(item, filename)
        # Load HDF5 file
        with h5py.File(file, 'r') as f:
            lambdas = np.array(f['lambdas'])
            transmission = np.array(f['transmission_right'])

        
        lambdas_indices = np.where(lambdas>0.4)
        min_trans = np.min(transmission[lambdas_indices])
        index_lambda_min_trans = np.where(transmission==min_trans)
        if True:
            transmission_mins += [min_trans]
            lambdas_mins += [lambdas[index_lambda_min_trans][0]]
            fig.add_trace(go.Scatter(x=[lambdas[index_lambda_min_trans][0]], y=[min_trans],
                                      mode='markers',showlegend =False),row=2,col=1)
        i+=1



      
# Update layout
fig.update_yaxes(title_text="T",type="log",showexponent = 'all',
                    exponentformat = 'e',row=2,col=1)
fig.update_xaxes(title_text="$\Large{\lambda(\mu m)}$",row=2,col=1)
fig.update_yaxes(title_text="$\Large{n,k}$",type="log",row=1,col=1)
fig.update_layout(
                
                font=dict(size=18),
              
                )

fig.add_trace(go.Scatter(x=data_refractive['wl'],y=data_refractive['n'],
                                      mode='lines',name="n"),row=1,col=1)

fig.add_trace(go.Scatter(x=data_abs['wl'],y=data_abs['k'],
                                      mode='lines',name="k"),row=1,col=1)

#Save the plot in PDF format
store_path = f"output/Plots_STL/Poster/Plots_absoption"
if not os.path.exists(store_path):
       os.makedirs(store_path)
       print(f"Folder '{store_path}' created successfully.")
     
fig.write_image(f"{store_path}/gap_position.png")
fig.show()
        
```

```python
#Gap position absorption plots

folder_path = ["output/Convergence_STL_Test_phi_Absorption/z_incidence/Data"]
sizes = [1.14,1.22,1.43,1.57,1.64,1.72,1.79]
transmission_mins=[]
lambdas_mins=[]
loss_position=[]
for j,item in enumerate(folder_path):
 i =0
 fig = go.Figure()
 for filename in sorted(os.listdir(item)):
    print(filename)
    if not Path(filename).suffix==".h5":
        continue 
    if os.path.isfile(os.path.join(item, filename)):
        file=os.path.join(item, filename)
        # Load HDF5 file
        with h5py.File(file, 'r') as f:
            lambdas = np.array(f['lambdas'])
            transmission = np.array(f['transmission_right'])
            one_minus_R = np.array(f['transmission_left'])
            loss = -(transmission-one_minus_R)

        
        lambdas_indices = np.where(lambdas>0.4)
        min_trans = np.min(transmission[lambdas_indices])
        index_lambda_min_trans = np.where(transmission==min_trans)
        if True:
            loss_position += [loss[index_lambda_min_trans][0]]
            lambdas_mins += [lambdas[index_lambda_min_trans][0]]
            fig.add_trace(go.Scatter(x=[lambdas[index_lambda_min_trans][0]], y=([loss[index_lambda_min_trans][0]]),
                                      mode='markers', name=f"edge={sizes[i]}um"))
        i+=1



      
# Update layout
fig.update_layout(
                title=dict(text="$\\text{Gap Position vs absorption}$", font=dict(size=18)),
                xaxis_title='$\lambda (\mu\\text{m})$',
                yaxis_title='Absorption',
                yaxis_gridcolor='lightgrey',  # Set grid color
                plot_bgcolor='rgba(0,0,0,0)',  # Set background color of plot,
                yaxis_range=[0, 0.11]  # Set the range for y-axis
                
                )

#Save the plot in PDF format
store_path = f"output/Plots_STL/Absorption"
if not os.path.exists(store_path):
       os.makedirs(store_path)
       print(f"Folder '{store_path}' created successfully.")
     
#fig.write_image(f"{store_path}/loss_gap_position.pdf")
fig.show()
        
```
