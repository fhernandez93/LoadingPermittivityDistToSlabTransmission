```python
import os
from pathlib import Path
import re
import shutil
```

```python
directory = r"D:\MPB Code and structures\20250731 New Structures"
```

```python
# Template for file.ctl
ctl_template = """;################  H5 FILE WITH THE EPSILON  ################ 
(set! eigensolver-block-size -257)
(set-param! mesh-size 7)
; OUT YOUR INPUT H5 FILE HERE
(set! epsilon-input-file "{h5file}") 

; generating basis 
(set! geometry-lattice (make lattice (basis-size  1.0       1.0       1.0      ) 
                                     (basis1    1.0       0.0       0.0       ) 
                                     (basis2    0.0       1.0       0.0       ) 
                                     (basis3    0.0       0.0       1.0       ))) 

; we'll calculate only at the Gamma point to start with

(define Gamma (vector3 0 0 0))
(define R (vector3 0.5 0.5 0.5))
(define X (vector3 0 0.5 0))
(define M (vector3 0.5 0.5 0))
(define Mp (vector3 0.5 0 0.5))
(define Xp (vector3 0 0 0.5))
(define-param k-interp 1)

;Quite comprehensive path in ths SC lattice
;(set! k-points (interpolate k-interp (list Gamma X M  R Gamma M X R)))

;NOTICE WE'RE JUST CALCULATING AT GAMMA
;Quick calculation only at Gamma
(set! k-points (interpolate k-interp (list Gamma X)))

(set-param! resolution 256)
(set! tolerance 1e-6)
(set-param! num-bands 600)
(set! output-epsilon (lambda () (print "skipping output-epsilon\\n")))
(run)
"""
```

```python
for dirpath, dirnames, filenames in os.walk(directory):
    try:
        n_value = float(re.search(r'n_([+-]?\d+(?:\.\d+)?)', dirpath).group(1))
        perp = float(re.search(r'perp_([+-]?\d+(?:\.\d+)?)', dirpath).group(1))
        if perp==0.4:
            dir_new = rf"{directory}/20250731 Structures_For_Marian_new/Ovalloidal_cylinder/index_{n_value}"
        else:
            dir_new = rf"{directory}/20250731 Structures_For_Marian_new/Perfect_Cylinder/index_{n_value}"
        
        os.makedirs(dir_new,exist_ok=True)
        for file in filenames:
            if file.endswith(".h5") and "eps" in file:
                # src = os.path.join(dirpath, file)
                # dst = os.path.join(dir_new, file)
                # shutil.copy2(src, dst)
                ctl_content = ctl_template.format(h5file=file)
                ctl_path = os.path.join(dir_new, f"{Path(file).stem}.ctl")
                with open(ctl_path, "w") as f:
                    f.write(ctl_content)

    except Exception as e:
        # Optional: print skipped dirs for debugging
        print(f"Skipped {dirpath}: {e}")
        continue
```
